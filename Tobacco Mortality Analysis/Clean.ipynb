```python
import pandas as pd
from functools import reduce

# Load CSV files
prescriptions = pd.read_csv("prescriptions.csv")
metrics = pd.read_csv("metrics.csv")
fatalities = pd.read_csv("fatalities.csv")
smokers = pd.read_csv("smokers.csv")
admissions = pd.read_csv("admissions.csv")

# Function to clean 'Year' column
def clean_year(year_val):
    try:
        return int(str(year_val)[:4])  # e.g., "2014/15" → 2014
    except:
        return None

# Standardize 'Year' column in each DataFrame
for df in [prescriptions, metrics, fatalities, smokers, admissions]:
    df["Year"] = df["Year"].apply(clean_year)

# Merge all datasets using outer join on 'Year'
dfs = [prescriptions, metrics, fatalities, smokers, admissions]
master_df = reduce(lambda left, right: pd.merge(left, right, on="Year", how="outer"), dfs)

# Inspect merged data
print(master_df.shape)
print(master_df.isnull().sum().sort_values(ascending=False).head(20))

# Drop rows with any missing values
master_df = master_df.dropna()

# Save the cleaned dataset
master_df.to_csv("tobacco_mortality.csv", index=False)
print("Cleaned dataset saved as tobacco_mortality.csv")
```
